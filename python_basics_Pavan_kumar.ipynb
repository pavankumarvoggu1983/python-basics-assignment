```python
Age = int(input("Enter  your age: "))
Age_months = Age * 12
print(f"Your age in months is: {Age_months}")
Age_days = Age * 12 * 365
print(f"Your age in days is: {Age_days}")
Age_reminder_7 = Age % 7
print(f"remainder when age is divided by 7: {Age_reminder_7}")
Age_raisedpower2 = Age ** 2
print(f"Age raised to the power of 2 is: {Age_raisedpower2}")
```
